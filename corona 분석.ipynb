```python
import urllib.request
import xml.dom.minidom
from xml.etree import ElementTree as ET
import pandas as pd

encodingKey = "%2Fh0Up1f99lWuK7xcxzJ3Qz6xjorpXVNBej4VwlxCyhlD8Gqe11SIu68n1YK81cZT6KRAZxez0cuM%2B31Hbl7KTg%3D%3D"
decodingKey = "/h0Up1f99lWuK7xcxzJ3Qz6xjorpXVNBej4VwlxCyhlD8Gqe11SIu68n1YK81cZT6KRAZxez0cuM+31Hbl7KTg=="

def getData(numOfRows, pageNO, startCreateDt, endCreateDt) :
    # request url정의
    url = (
        "http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?serviceKey=" + str(encodingKey) + 
        "&pageNO=" + str(pageNO)+ 
        "&numOfRows=" + str(numOfRows) + 
        "&startCreateDt=" + str(startCreateDt) + 
        "&endCreateDt=" + str(endCreateDt)
    )
    
    request = urllib.request.Request(url)

    # request보내기
    response = urllib.request.urlopen(request)

    # 결과 코드 정의
    rescode = response.getcode()

    if(rescode==200):
        response_body = response.read()
        dom = xml.dom.minidom.parseString(response_body.decode('utf-8'))
        pretty_xml_as_string = dom.toprettyxml()
        print(pretty_xml_as_string)
    else:
        print("Error Code:" + rescode)

#     print(url)
    
    return response_body

def parse1(response_body):
    root_element = ET.fromstring(response_body) # 문자열에서 XML을 파싱합니다
    rows = [] # item 저장할 list 초기화한다
    iter_element = root_element.iter(tag="item") # item 태그 iterator를 가져옵니다
    for element in iter_element: # item 태그를 순회합니다
        row = {} # 각 item을 저장할 dict 초기화한다
        row['clearCnt'] = element.find("clearCnt").text # clearCnt 값을 저장합니다
        row['createDt'] = element.find("createDt").text # createDt 값을 저장합니다
        row['deathCnt'] = element.find("deathCnt").text # deathCnt 값을 저장합니다
        row['examCnt'] = element.find("examCnt").text # examCnt 값을 저장합니다
        row['seq'] = element.find("seq").text # seq 값을 저장합니다
        row['stateDt'] = element.find("stateDt").text # stateDt 값을 저장합니다
        row['stateTime'] = element.find("stateTime").text # stateTime 값을 저장합니다   
        row['updateDt'] = element.find("updateDt").text # updateDt 값을 저장합니다   

        rows.append(row) # itmes에 item을 저장합니다
#     print(rows) # 결과를 출력한다

    columns = ["clearCnt", "createDt", "deathCnt", "examCnt", "seq", "stateDt", "stateTime", "updateDt"]
    df = pd.DataFrame(rows, columns = columns)
    return(df)
    
    
def parse2(response_body):
    root_element = ET.fromstring(response_body) # 문자열에서 XML을 파싱합니다
    rows = [] # item 저장할 list 초기화한다
    iter_element = root_element.iter(tag="item") # item 태그 iterator를 가져옵니다
    for element in iter_element: # item 태그를 순회합니다
        row = {} # 각 item을 저장할 dict 초기화한다
        row['accDefRate'] = element.find("accDefRate").text # accDefRate 값을 저장합니다
        row['accExamCnt'] = element.find("accExamCnt").text # accExamCnt 값을 저장합니다
        row['accExamCompCnt'] = element.find("accExamCompCnt").text # accExamCompCnt 값을 저장합니다
        row['clearCnt'] = element.find("clearCnt").text # clearCnt 값을 저장합니다
        row['createDt'] = element.find("createDt").text # createDt 값을 저장합니다
        row['deathCnt'] = element.find("deathCnt").text # deathCnt 값을 저장합니다
        row['examCnt'] = element.find("examCnt").text # examCnt 값을 저장합니다
        row['resutlNegCnt'] = element.find("resutlNegCnt").text # resutlNegCnt 값을 저장합니다
        row['seq'] = element.find("seq").text # seq 값을 저장합니다
        row['stateDt'] = element.find("stateDt").text # stateDt 값을 저장합니다
        row['stateTime'] = element.find("stateTime").text # stateTime 값을 저장합니다   
        row['updateDt'] = element.find("updateDt").text # updateDt 값을 저장합니다   

        rows.append(row) # itmes에 item을 저장합니다
#     print(rows) # 결과를 출력한다

    columns = ["accDefRate", "accExamCnt", "accExamCompCnt", "clearCnt", "createDt", "deathCnt", "examCnt", "resutlNegCnt", "seq", "stateDt", "stateTime", "updateDt"]
    df = pd.DataFrame(rows, columns = columns)
    
    return(df)
    
def parse3(response_body):
    root_element = ET.fromstring(response_body) # 문자열에서 XML을 파싱합니다
    rows = [] # item 저장할 list 초기화한다
    iter_element = root_element.iter(tag="item") # item 태그 iterator를 가져옵니다
    for element in iter_element: # item 태그를 순회합니다
        row = {} # 각 item을 저장할 dict 초기화한다
        row['accDefRate'] = element.find("accDefRate").text # accDefRate 값을 저장합니다
        row['accExamCnt'] = element.find("accExamCnt").text # accExamCnt 값을 저장합니다
        row['accExamCompCnt'] = element.find("accExamCompCnt").text # accExamCompCnt 값을 저장합니다
        row['careCnt'] = element.find("careCnt").text # careCnt 값을 저장합니다
        row['clearCnt'] = element.find("clearCnt").text # clearCnt 값을 저장합니다
        row['createDt'] = element.find("createDt").text # createDt 값을 저장합니다
        row['deathCnt'] = element.find("deathCnt").text # deathCnt 값을 저장합니다
        row['examCnt'] = element.find("examCnt").text # examCnt 값을 저장합니다
        row['resutlNegCnt'] = element.find("resutlNegCnt").text # resutlNegCnt 값을 저장합니다
        row['seq'] = element.find("seq").text # seq 값을 저장합니다
        row['stateDt'] = element.find("stateDt").text # stateDt 값을 저장합니다
        row['stateTime'] = element.find("stateTime").text # stateTime 값을 저장합니다   
        row['updateDt'] = element.find("updateDt").text # updateDt 값을 저장합니다   

        rows.append(row) # itmes에 item을 저장합니다
#     print(rows) # 결과를 출력한다

    columns = ["accDefRate", "accExamCnt", "accExamCompCnt", "careCnt", "clearCnt", "createDt", "deathCnt", "examCnt", "resutlNegCnt", "seq", "stateDt", "stateTime", "updateDt"]
    df = pd.DataFrame(rows, columns = columns)
    
    return(df)
```
